# L03.0 Basic plots with Matplotlib, Plotly, and Bokeh 

## Matplotlib

Matplotlib is arguably the most popular Python package for both basic and advanced plotting. Let's look at a basic plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)

plt.plot(x, y)
plt.show()
```

This looks pretty neat already. All information is clearly visible. However, we can add a few tweaks to make it more visually appealing. Let's use the golden ratio (1.618) as the figure's aspect ratio and up its resolution to 600 dpi. We of course also need to annotate the axes with labels.

```python
golden = 1.618
h = 2.5; w = h*golden 
fig = plt.figure(figsize=(w,h), dpi=600)

plt.plot(x, y, lw=2)
plt.xlabel("x")
plt.ylabel("cos(x)")

fig.tight_layout()
plt.show()
```

## Plotyly

The same plot in Plotly

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines'))
fig.update_layout(xaxis_title='x', yaxis_title='cos(x)')
fig.show()
```

Plotly creates more than just a plot: It creates an interactive widget. The default layout is based on being able to easily read data directly. We can easily customise our plot to make it look more like the Matplotlib version

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(width=2)))
fig.update_layout(xaxis=dict(
        title="x",
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
    ),
    yaxis=dict(
        title="cos(x)",
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
    ),
    autosize=False,
    plot_bgcolor='white',
)
fig.show()
```

## Bokeh

The same plot in Bokeh

```python
from bokeh.plotting import figure, output_file, show

output_file("basic_bokeh.html")

p = figure()
p.line(x, y)

show(p)
```

Running this code creates a html file and opens it in your browser.

```python
output_file("basic_bokeh_pretty.html")

golden = 1.618
dpi = 75
h = int(2.5*dpi); w = int(h*golden) 

p = figure(plot_width=w, plot_height=h)
p.line(x, y, line_width=2)

p.xaxis.axis_label = "x"
p.yaxis.axis_label = "cos(x)"

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.xaxis.minor_tick_line_color = None
p.yaxis.minor_tick_line_color = None

show(p)
```
